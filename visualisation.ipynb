```python
import gwaslab as gl
import pandas as pd

loci_path = "/home/wcy/data/UKB/eye_feature/disease_cor/loci_file/left_flitter_dim64_pheno_result_loci.txt"
save_path = "/home/wcy/data/UKB/eye_feature/disease_cor/novel_loci/left_flitter-SLE.csv"
efo_id = "MONDO_0007915"

gwas_data =pd.read_csv(loci_path,sep=",")
mysumstats = gl.Sumstats(gwas_data,
             snpid="SNPID",
             chrom="CHR",
             pos="POS",
             ea="EA",
             nea="NEA",
             n="N",
             p="P", 
             build="19",
             verbose=False)

mysumstats.basic_check()
mysumstats.data["CHR"] = mysumstats.data["CHR"].astype(str).str.replace("chr", "").astype(int)
mysumstats_hg38 = mysumstats.liftover(n_cores=3, from_build="19", to_build="38")
result = mysumstats.get_novel(efo=efo_id)
result = result.rename(columns={'POS': 'POS_hg38'})
result = pd.merge(gwas_data[['SNPID', 'POS']], result ,on="SNPID")
result.to_csv(save_path,sep="\t",index=False)

filtered_result = result[result['NOVEL'] == False]
filtered_result
```

# 数据分布

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 读取数据
df = pd.read_csv('/home/wcy/data/UKB/eye_feature/feature_data/left_vessel_dim128.csv')

# 创建8×8的图表
fig, axes = plt.subplots(8, 8, figsize=(20, 25))
fig.suptitle('Distribution of 128 Latent Variables', y=1.02, fontsize=16)

# 扁平化axes数组以便循环
axes_flat = axes.flatten()

# 遍历64个潜在变量
for i in range(64):
    # 计算均值
    mean_val = df[f'latent_{i+1}'].mean()
    
    # 绘制直方图
    axes_flat[i].hist(df[f'latent_{i+1}'], bins=50, density=True)
    
    # 添加均值的垂直线
    axes_flat[i].axvline(x=mean_val, color='red', linestyle='--', linewidth=2)
    
    # 在图上标注均值
    axes_flat[i].text(0.98, 0.95, f'Mean: {mean_val:.2f}', 
                     transform=axes_flat[i].transAxes,
                     horizontalalignment='right',
                     verticalalignment='top',
                     fontsize=8,
                     color='red')
    
    axes_flat[i].set_title(f'Latent {i+1}')
    axes_flat[i].tick_params(labelsize=8)
    axes_flat[i].set_xlabel('')
    axes_flat[i].set_ylabel('')

plt.tight_layout()
plt.savefig('/home/wcy/data/UKB/eye_feature/disease_cor/pic_file/left_vessel_dim128_0417.png', dpi=300, bbox_inches='tight')
plt.show()
```
